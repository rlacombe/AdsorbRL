```python
def sample_trajectory(SARSlist, max_path_length, ob_terminal):
  '''
  args:
    SARSlist: list of possible (ob, ac, rew, next_ob) combinations, where:
      ob: 1-D np.array of length obs_dim
      ac: 1-D np.array of length acs_dim
      rew: float 
      next_ob: 1-D np.array of length obs_dim
    max_path_length: int max length of path
    ob_terminal: the terminal state (typically the min energy state in observations)

  returns:
    Path object 
  '''
  SARSlist_len = len(SARSlist)
  obs, acs, rewards, next_obs, terminals, image_obs = [], [], [], [], [], []
  steps = 0

  k0 = np.random.choice(SARSlist_len)
  (ob, ac, rew, next_ob) = SARSlist[k0]
  while True:
    obs.append(ob)
    acs.append(ac)
    next_obs.append(next_ob)
    rewards.append(rew)
    done = np.array_equal(next_ob, ob_terminal)
    steps += 1

    # If the episode ended, the corresponding terminal value is 1
    # otherwise, it is 0
    if done or steps > max_path_length:
        terminals.append(1)
        break
    else:
        terminals.append(0)

    # If not ended, sample at random a transition starting from the next observation
    possible_next_SARS = [sars for sars in SARSlist if np.array_equal(sars[0], next_ob)]
    (ob, ac, rew, next_ob) = random.choice(possible_next_SARS)

  return Path(obs, [], acs, rewards, next_obs, terminals)


def sample_trajectories(SARSlist, max_path_length, ob_terminal, min_timesteps_per_batch):

    timesteps_this_batch = 0
    paths = []
    while timesteps_this_batch < min_timesteps_per_batch:

        #collect rollout
        path = sample_trajectory(SARSlist, max_path_length, ob_terminal)
        paths.append(path)

        #count steps
        timesteps_this_batch += get_pathlength(path)
        print('At timestep:    ', timesteps_this_batch, '/', min_timesteps_per_batch, end='\r')

    return paths, timesteps_this_batch


def sample_n_trajectories(SARSlist, max_path_length, ob_terminal, ntraj):

    paths = []
    for i in range(ntraj):
        # collect rollout
        path = sample_trajectory(SARSlist, max_path_length, ob_terminal)
        paths.append(path)

    return paths


def Path(obs, image_obs, acs, rewards, next_obs, terminals):
    """
        Take info (separate arrays) from a single rollout
        and return it in a single dictionary
    """
    if image_obs != []:
        image_obs = np.stack(image_obs, axis=0)
    return {"observation" : np.array(obs, dtype=np.float32),
            "image_obs" : np.array(image_obs, dtype=np.uint8),
            "reward" : np.array(rewards, dtype=np.float32),
            "action" : np.array(acs, dtype=np.float32),
            "next_observation": np.array(next_obs, dtype=np.float32),
            "terminal": np.array(terminals, dtype=np.float32)}
```

#### Create the numpy matrix of states

```python
ns = len(starOH2element)
elems = [
    'Ag', 'Al',	'As', 'Au', 'B', 'Bi', 'C', 'Ca', 'Cd', 'Cl', 'Co', 'Cr', 'Cs', 'Cu', 'Fe', 'Ga', 'Ge', 'H', 'Hf', 'Hg', 'In', 'Ir', 'K', 'Mn', 'Mo', 'N', 'Na', 'Nb', 'Ni', 'Os', 'P', 'Pb', 'Pd', 'Pt', 'Rb', 'Re', 'Rh', 'Ru', 'S', 'Sb', 'Sc', 'Se', 'Si', 'Sn', 'Sr', 'Ta', 'Tc', 'Te', 'Ti', 'Tl', 'V', 'W', 'Y', 'Zn', 'Zr'
    ]
ne = len(elems)
states = np.zeros((ns, ne))
rewards = np.zeros(ns)

for i in range(ns):
  entry = starOH2element[i]
  rewards[i] = entry["Energy"]
  for j in range(ne):
    if entry[elems[j]] == '1':
      states[i, j] = 1
```

#### Generate the set of all possible tuples (s, a, r, s')
each tuple will be of the form: 
( np.array(ne), np.array(ne), float, np.array(ne) )

```python
na = ne
def get_neighbors(current_state):
  '''
  Gets a list of possible neighbors to a given state that an agent can 
  move to. 
  This means we can swap out on element, add one element, or delete one element 
  at a time but not to exceed 3 elements and must have at least one element. 
  The condition not to exceed 3 elements will be self-enforced because we only 
  allow s1 to come from the set S (so don't need to check the condition a second time).
  arguments:
    current_state: np.array of shape (ne, 1)
  returns:
    ix_neighbor: indexes of neighbor states
    actions: list of np.array, each of shape (ne, 1)
  '''
  current_ix = np.where(np.all(states == current_state, axis=1))[0][0]
  ix_neighbors = []
  # adds
  for i in range(ne):
    add_state = current_state.copy()
    add_state[i] += 1
    if list(add_state) in states.tolist():
      ix_neighbors.append(np.where(np.all(states == add_state, axis=1))[0][0])
  # subtracts
  for i in range(ne):
    sub_state = current_state.copy()
    sub_state[i] -= 1
    if list(sub_state) in states.tolist():
      ix_neighbors.append(np.where(np.all(states == sub_state, axis=1))[0][0])
  # exchanges
  for i in range(ne):
    if current_state[i] == 1: 
      exch_state = current_state.copy()
      exch_state[i] = 0
      for j in range(ne):
        if j == i:
           continue
        exch_state2 = exch_state.copy()
        exch_state2[j] +=1 
        if list(exch_state2) in states.tolist():
          ix_neighbors.append(np.where(np.all(states == exch_state2, axis=1))[0][0])
  # print(ix_neighbors)
  actions = states[ix_neighbors, :] - current_state
  return ix_neighbors, actions


ix_neighbors, actions = get_neighbors(states[2109, :])
print("num neighbors = ", len(ix_neighbors))
print(ix_neighbors)
print(actions)
```
